```python
#/usr/bin/env python3

# pip install periodictable
# pip install google-cloud-translate==2.0.1
import periodictable
import six
from google.cloud import translate_v2
import sys
# from ruamel.yaml import YAML
import ruamel.yaml
import pandas as pd
```

```python
# requires a google cloud account, service account, credentials, permission to use project in translator
def translate_text_with_model(target, text, model="nmt"):
    translate_client = translate_v2.Client()
    if isinstance(text, six.binary_type):
        text = text.decode("utf-8")
    # Text can also be a sequence of strings, in which case this method
    # will return a sequence of results for each text.
    result = translate_client.translate(text, target_language=target, model=model)
    return result["translatedText"]
```

```python
en = []
es = []
fr = []
ar = []
zh = []
ru = []
data = []
for el in periodictable.elements:
    en.append(el.name)
    es.append(translate_text_with_model(target='es', text=el.name))
    fr.append(translate_text_with_model(target='fr', text=el.name))
    ar.append(translate_text_with_model(target='ar', text=el.name))
    zh.append(translate_text_with_model(target='zh', text=el.name))
    ru.append(translate_text_with_model(target='ru', text=el.name))
    print(el.name, es[-1], fr[-1], ar[-1], zh[-1], ru[-1])
df = pd.DataFrame({'en': en, 'es': es, 'fr': fr, 'ar': ar, 'zh': zh,'ru': ru})
print(df)
df.to_csv('out.csv', index=False)
```

```python
def flow_mode(x):
    # https://stackoverflow.com/questions/56937691/making-yaml-ruamel-yaml-always-dump-lists-inline
    retval = ruamel.yaml.comments.CommentedSeq(x)
    retval.fa.set_flow_style()  # fa -> format attribute
    return retval

lang = [{"english": en}, 
        {"spanish": es}, 
        {"french": fr}, 
        {"arabic": ar}, 
        {"chinese": zh}, 
        {"russian": ru}]

for x in lang:
    for k, v in x.items():
        print("processing ", k)
        # an existing names.yml must exist in the dir this is run in. it uses that as the template to replace
        file_name = 'names.yml'
        outfile = k + '-' + 'names.yml'
        # name files generated should be copied into src/main/resources when done
        yaml = ruamel.yaml.YAML()

        with open(file_name) as fp:
            data = yaml.load(fp)

        data['english']['M'] = flow_mode(v)
        data['english']['F'] = flow_mode(v)
        data['english']['family'] = flow_mode(v)
        data ['spanish']['M'] = flow_mode(v)
        data ['spanish']['F'] = flow_mode(v)
        data ['spanish']['family'] = flow_mode(v)

        with open(outfile, 'w') as fp:
            yaml.dump(data, fp)

        # display file
        with open(outfile) as fp:
            print(fp.read(), end='')        
```
